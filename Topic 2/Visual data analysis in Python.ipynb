```python
tsne=TSNE(random_state=17)
```

```python
%%time
X_repr=tsne.fit_transform(df)
```

```python
X_repr.shape
```

```python
plt.rcParams['figure.figsize'] = (8,6)
plt.scatter(X_repr[df['Churn']==1,0], X_repr[df['Churn']==1,1], alpha=.5, c='blue',
           label='churn');
plt.scatter(X_repr[df['Churn']==0,0], X_repr[df['Churn']==0,1], alpha=.5, c='orange',
           label='loyal');
plt.xlabel('tsne axis #1')
plt.ylabel('tsne axis #2')
plt.legend()
plt.title('tSNE representation')
plt.savefig('churn_tsne.png', dpi=300);
```
